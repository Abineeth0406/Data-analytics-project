```python
# my dataset project
import pandas as pd
import seaborn as sns
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder, StandardScaler

# Load dataset
dataset = pd.read_csv('/content/creditdataset.csv')

# Fill missing values
dataset[['Saving_acc', 'Checking_acc']] = dataset[['Saving_acc', 'Checking_acc']].fillna('unknown')

# Correlation heatmap for numeric fields only
numeric_cols = dataset.select_dtypes(include=['int64', 'float64']).columns
sns.heatmap(dataset[numeric_cols].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()

# Boxplot for Credit amount by Housing status
sns.boxplot(x='Housing', y='Credit_amount', data=dataset, palette='Set2')
plt.title('Credit Amount by Housing Status')
plt.show()

# Encode categorical features and scale numeric features
categorical_cols = ['Sex', 'Housing', 'Saving_acc', 'Checking_acc', 'Purpose', 'Employment_Status']
for col in categorical_cols:
    dataset[col] = LabelEncoder().fit_transform(dataset[col].astype(str))

features = ['Age', 'Job', 'Credit_amount', 'Duration']
dataset_scaled = StandardScaler().fit_transform(dataset[features])

# Apply K-Means clustering
dataset['Cluster'] = KMeans(n_clusters=3, random_state=42).fit_predict(dataset_scaled)

# Visualize clusters
sns.scatterplot(x=dataset['Age'], y=dataset['Credit_amount'], hue=dataset['Cluster'], palette='viridis')
plt.title('Customer Clusters')
plt.xlabel('Age')
plt.ylabel('Credit Amount')
plt.show()

# Output cluster counts and save data
print(dataset['Cluster'].value_counts())
dataset[['Age', 'Job', 'Housing', 'Saving_acc', 'Checking_acc', 'Credit_amount', 'Cluster']].to_csv('grouped_customers.csv', index=False)
```
