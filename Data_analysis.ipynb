# HEAT-MAP CORRELATION(FIXED)

This code computes the correlation matrix for numeric columns in the dataset and visualizes it as a heatmap using Seaborn. It helps identify correlations between variables, with stronger correlations shown in darker colors, and annotates the correlation values within the heatmap...

```python
# my dataset project
import pandas as pd
import seaborn as sns
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder, StandardScaler

# Load dataset
dataset = pd.read_csv('/content/creditdataset.csv')

# Fill missing values
dataset[['Saving_acc', 'Checking_acc']] = dataset[['Saving_acc', 'Checking_acc']].fillna('unknown')

# Correlation heatmap for numeric fields only
numeric_cols = dataset.select_dtypes(include=['int64', 'float64']).columns
sns.heatmap(dataset[numeric_cols].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
dataset.shape
```

# BOXPLOT FOR CREDIT AMOUNT BY HOUSING DATA

This code generates a box plot to visualize the distribution of "Total Amount across different 'Product Category' values. The box plot highlights the median, quartiles, and potential outliers in the total amount for each product category.

```python
# Boxplot for Credit amount by Housing status
sns.boxplot(x='Housing', y='Credit_amount', data=dataset, palette='Set2')
plt.title('Credit Amount by Housing Status')
plt.show()
dataset.columns
```

# PIE-CHART FOR HOUSING STATUS

A pie chart is a circular graph that is divided into slices to illustrate numerical proportions. Each slice represents a category or a part of the whole, and its size is proportional to the percentage or fraction of that category in relation to the total. Here's how to understand and interpret a pie chart..

```python
# Pie chart for Housing status distribution
housing_counts = dataset['Housing'].value_counts()
plt.figure(figsize=(6, 6))
mylabels=["FREE","RENT","OWN"]
myexplode=[0.1,0.1,0.1]
housing_counts.plot.pie(autopct='%1.1f%%',labels=mylabels, startangle=120,explode=myexplode,shadow=True, cmap='Set3')
plt.title('Housing Status Distribution')
plt.ylabel('')  # Hides the y-axis label
plt.show()
```

# SCATTER PLOT FOR DURATION VS CREDIT AMOUNT

A scatter plot is a type of data visualization that uses dots to represent values for two variables commonly used to observe relationships, trends, and patterns between two continuous variables..

```python
# Scatter plot for Duration vs Credit Amount
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Duration', y='Credit_amount', data=dataset, hue='Saving_acc', palette='Set2')
plt.title('Duration vs Credit Amount')
plt.xlabel('Duration')
plt.ylabel('Credit Amount')
plt.legend(title='Saving Account')
plt.show()
```

# PAIR PLOT FOR NUMERIC FIELDS

A pair plot (also known as a scatterplot matrix) is a grid of scatter plots used to visualize relationships between multiple variables in a dataset. It displays scatter plots of all possible pairwise combinations of variables, which helps to assess the relationships and distributions between variables simultaneously..

```python
# Pairplot for numeric fields
sns.pairplot(dataset, vars=numeric_cols, hue='Housing', palette='Set2')
plt.suptitle('Pairplot for Numeric Features', y=1.02)
plt.show()
```

# GRAPHICAL REPRESENTATION:LOAN PURPOSE COUNT

A graphical representation of loan purpose count is a visualization that shows how many loans have been given for each specific purpose. This type of chart is helpful to understand the distribution and frequency of various loan purposes within a dataset. The most common types of graphs used to represent loan purpose count are bar charts or pie charts, as they allow easy comparison of the different categories..

```python
# Graphical representation: Loan Purpose counts
loan_purpose_counts = dataset['Purpose'].value_counts()
plt.figure(figsize=(10, 6))
loan_purpose_counts.plot(kind='line',marker='D',mec='r',ms=10,linewidth=3.5, color=sns.color_palette('muted'))
plt.title('Loan Purpose Distribution')
plt.xlabel('Purpose')
plt.ylabel('Count')
plt.show()
```

# LINE CHART FOR CREDIT AMOUNT AGE

The line chart shows the average credit amount for each age group, helping us identify patterns or trends in how credit amounts vary with age.

```python
# Line chart for Credit Amount across Age
age_credit = dataset.groupby('Age')['Credit_amount'].mean()
age_credit.plot(kind='line', color='darkgreen', marker='o')
plt.title('Average Credit Amount by Age')
plt.xlabel('Age')
plt.ylabel('Average Credit Amount')
plt.grid()
plt.show()
print(dataset.loc[0])#accessing column integer by index
```

# BAR CHART FOR JOB STATUS COUNT

A bar chart is a visual representation used to show the frequency or proportion of different categories within a variable. In this case, we are looking at the Job distribution or similar categorical data...

```python
# Bar chart for Job status counts
sns.barplot(x=dataset['Job'].value_counts().index, y=dataset['Job'].value_counts().values, palette='coolwarm')
plt.title('Job Distribution')
plt.xlabel('Job')
plt.ylabel('Count')
plt.show()
```

# HISTOGRAPHIC REPRESENTATION

A histogram is a graphical representation of the distribution of a single numerical variable. It organizes the data into intervals (bins) and counts how many data points fall into each interval...

```python
#Histogram: Distribution of Credit Amount
plt.hist(dataset['Credit_amount'], bins=20, color='darkred', edgecolor='black', alpha=0.7)
plt.title('Histogram of Credit Amount')
plt.xlabel('Credit Amount')
plt.ylabel('Frequency')
plt.show()
print(dataset.iloc[0])#accessing row by index
```

# VIOLINE-PLOT REPRESENTATION

A violin plot is a combination of a boxplot and a kernel density plot. It provides a visual summary of the distribution of data across different categories, offering insights into data distribution, spread, and skewness.

```python
#Violin plot for Credit Amount by Housing Status
if not dataset.empty and 'Housing' in dataset.columns and 'Credit_amount' in dataset.columns:
    sns.violinplot(x='Housing', y='Credit_amount', data=dataset, palette='muted')
    plt.title('Violin Plot of Credit Amount by Housing Status')
    plt.show()
else:
    print("Error: Dataset is empty or missing required columns.")
```
